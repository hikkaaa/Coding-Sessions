### 7 - Transposing
We can change the orientation of image using transpose function. Transpose function takes one parameter and is method of transpose. There are six transposing methods you can use. 

Image.FLIP_LEFT_RIGTH (This inverse the rows of image)
Image.FLIP_TOP_BOTTOM (This inverse the colums of image)
Image.ROTATE_90 (This rotates the image by angle 90)
Image.ROTATE_180 (This rotates the image by angle of 180)
Image.ROTATE_270 (This rotates the image by angle of 270)
Image.TRANSPOSE (This converts the rows into columns)

```python
transpose_image1 = image.transpose(Image.FLIP_LEFT_RIGHT)
transpose_image2 = image.transpose(Image.FLIP_TOP_BOTTOM)
transpose_image3 = image.transpose(Image.ROTATE_90)
transpose_image4 = image.transpose(Image.ROTATE_180)
transpose_image5 = image.transpose(Image.ROTATE_270)
transpose_image6 = image.transpose(Image.TRANSPOSE)

plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(transpose_image1)
plt.title('FLIP_LEFT_RIGHT')
plt.subplot(3,2,2)
plt.imshow(transpose_image2)
plt.title('FLIP_TOP_BOTTOM')
plt.subplot(3,2,3)
plt.imshow(transpose_image3)
plt.title('ROTATE_90')
plt.subplot(3,2,4)
plt.imshow(transpose_image4)
plt.title('ROTATE_180')
plt.subplot(3,2,5)
plt.imshow(transpose_image5)
plt.title('ROTATE_270')
plt.subplot(3,2,6)
plt.imshow(transpose_image6)
plt.title('TRANSPOSE')
```

### 8 - Resizing
This function changes the size of Image in terms of width and heigth.\
This function accepts a dual tuple pair that is the new size of Image.\
This returns the image object. \
Reducing or increasing the size involves various algorithms known as interpolations. 

# Interpolation Techniques

___

**NEAREST** \
Pick one nearest pixel from the input image. Ignore all other input pixels. Simple technique, results are king of poor. 

**BOX** \
Each pixel of source image contributes to one pixel of the destination image with identical weights. For upscaling is equivalent of NEAREST. 

**BILINEAR**

For resize calculate the output pixel value using linear interpolation on all pixels that may contribute to the output value. 

**HAMMING** \
Produce a sharper image that BILINEAR, does not have dislocations on local level like with BOX.

**BICUBIC** \
One of the better ones; more expensive time-wise.

**LANCZOS** \
One of the better ones; more expensive time-wise.

```python
newsize = (300, 300)
plt.figure(figsize=(10,10))

resized_image1 = image.resize(newsize, Image.BILINEAR)
resized_image2 = image.resize(newsize, Image.NEAREST)
resized_image3 = image.resize(newsize, Image.BOX)
resized_image4 = image.resize(newsize, Image.HAMMING)
resized_image5 = image.resize(newsize, Image.BICUBIC)
resized_image6 = image.resize(newsize, Image.LANCZOS)

plt.subplot(3,2,1)
plt.imshow(resized_image1)
plt.title('BILINEAR')
plt.subplot(3,2,2)
plt.imshow(resized_image2)
plt.title('NEAREST')
plt.subplot(3,2,3)
plt.imshow(resized_image3)
plt.title('BOX')
plt.subplot(3,2,4)
plt.imshow(resized_image4)
plt.title('HAMMING')
plt.subplot(3,2,5)
plt.imshow(resized_image5)
plt.title('BICUBIC')
plt.subplot(3,2,6)
plt.imshow(resized_image6)
plt.title('LANCZOS')
```

### 15 - Image Enhancement
Image enhancement is increasing the pixels quality so that image result is more suitable to be used. There are different kinds of image enhancement techniques in image progressing. They include color, contrast, sharpness, brightness, etc.
#### Color Enhancement
Image colors are enhanced. 
#### Sharpness Enhancement
Contrast between dark and bright regions are increased.
#### Contrast Enhancement
Degree of color or grayscale variation are increased.
#### Brightness Enhancement
Overall image darkness is enhanced. 
All the functions color, contrast, sharpness and brightness are of Image Enhanced Module of Image Library. There functions take image as parameter and return the converted image. We can enhance images depending upon over need. There is a function enhance that takes float number that acts as a factor. Factor value varies from 0 to 1.0 means no change and 1 means maximum enhancement. 

```python
from PIL import ImageEnhance
plt.figure(figsize = (10,10))
image_color_enhan = image.copy()
_image1 = ImageEnhance.Color(image_color_enhan).enhance(2.5)
_image2 = ImageEnhance.Contrast(image_color_enhan).enhance(2.5)
_image3 = ImageEnhance.Brightness(image_color_enhan).enhance(1.5)
_image4 = ImageEnhance.Sharpness(image_color_enhan).enhance(2.5)

plt.subplot(2,2,1)
plt.imshow(_image1)
plt.title('Color')
plt.subplot(2,2,2)
plt.imshow(_image2)
plt.title('Contrast')
plt.subplot(2,2,3)
plt.imshow(_image3)
plt.title('Brightness')
plt.subplot(2,2,4)
plt.imshow(_image4)
plt.title('Sharpness')
```

### 16 - Alpha Blending
Alpha Blending is the mixing of the two images. It is combining of the alpha channel with other layers in an image in order to show translucency.

out = image1 * (1.0-alpha) + image2 * alpha

If the aplha is 0.0, a copy of the first image is returned. If the alpha is 1.0, a copy of the second image is returned. There are no restrictions on the alpha value. If necessary, the result is clipped to fit into the allowed output range (0 and 1),

There are two rules in order to blend images:\
**1-** Make sure the images have the same size\
**2-** Make sure the images have the alpha channel (* .png format)

The function **blend** of Image module takes three parameters. First two take the images to blend and third parameter takes the alpha number. It varies between 0 and 1. This number responsible for varying the translucency between two images.

```python
first = image2.copy()
second = image3.copy()

second = second.resize(first.size)

image_blend = Image.blend(first, second, 0.3)
plt.imshow(image_blend)
```

### 17 - Image Transform 
Image transform can be applied to an image to convert it from one domain to another. There are different types of transformation in python image library. They all come under a module ImageTransform. 

- **AffineTransform**: it preserves straight line, points and planes. The parallel line/rows same after the tranformation. Data is a 6-tuple (a,b,c,d,e,f) which contains the first two rows from an affine transform matrix. For each pixel (x,y) in the output image, the new value is taken from a position (a x + b y + c, d x + e y + f) in the input image, rounded to the nearest pixel. This function can be used to scale, translate, rotate, and shear the original image. 


- **ExtentTransform**: it computes the minimum and maximum values for a data field, producing a [min, max] array. This transform is useful for computing a value range and binding it to a signal name.


- **QUAD**: map a quadrilateral to a rectangle


- **MESH**: map a number of source quadrilaterals in one operation

In PIL there is a function transform that takes four parameters to transform the image:
1. size of image
2. type of transformation
3. matrix value for transformation (6-tuples values of Affine and 4-Tuple values for Extent
4. interpolation (optional)

```python
image_transform = image.copy()
image_transform = image_transform.transform(image_transform.size, Image.AFFINE,(1, -0.4, 0.2*image_transform.size[0],0,1,0))
plt.imshow(image_transform)
```

```python
image_transform = image.copy()
image_transform = image_transform.transform(image_transform.size, Image.EXTENT,(100,100,image_transform.size[0],image_transform.size[1]//2))
plt.imshow(image_transform)

```

### 18 - Flipping Channels
We can do the flipping of channels by getting the channels first and then merging them in your desired orientation. Split function is used to return the channels of image. Merge takes two parameters, orientations and the values of of channels. It returns the image so that we can use it further. 

```python
image_channels = image3.copy()
r, g, b = image_channels.split()
im = Image.merge('RGB', (b,g,r))
plt.imshow(im)
```
